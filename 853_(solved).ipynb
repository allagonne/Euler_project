```python
def pisano_period(modulo):
  return_list = [1, 1]
  while True:
    return_list.append((return_list[-2] + return_list[-1]) % modulo)
    if return_list[-2] == 1 and return_list[-1] == 1:
      period = len(return_list)-2
      return period
    elif len(return_list) > 122:
      return None
count = 0
max = 10**9+1
for i in range(3, max):
  if pisano_period(i) == 120:
    print(i,pisano_period(i))
    count += i
count
```

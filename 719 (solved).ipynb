```python
def splitter(n):
    s = str(n)
    for i in range(1, len(s)):
        start = s[0:i]
        end = s[i:]
        #display(start, end)
        yield [start, end]
        for split in splitter(end):
            #print(split)
            result = [start]
            result.extend(split)
            yield result

#combinations = list(splitter(7896))
#print(combinations)
def is_snum(nsq, n):
    combis = list(splitter(nsq))
    combis = [[int(i) for i in combi] for combi in combis]
    #print(combis)
    for combi in combis:
        #display(nsq, combi, sum(combi))
        if sum(combi) == n:
            print('succeed', nsq)
            return True
    return False
def get_snums(maxi = 10**4):
    i = 4 
    isq = i**2
    mylist = []
    while isq<maxi:
        isq = i**2
        if is_snum(isq, i):
            mylist.append(isq)
        i+=1
    return mylist
def get_sum_snums(maxi = 10**4):
    return sum(get_snums(maxi))
get_sum_snums(10**12)
```
